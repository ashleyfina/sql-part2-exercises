To get started with the exercises, go ahead and take a look at RideShareDB. This database was created with data from a dataset on [Kaggle](https://www.kaggle.com/fivethirtyeight/uber-pickups-in-new-york-city). The dataset was made using data obtained from the NYC Taxi and Limosine Commission or TLC. TLC tracks this data by base and even rideshare companies such as Uber and Lyft have assigned bases that each pickup must be attributed to. These tables do not include any data from Uber.

## String Functions

Write a query that returns the last 3 characters of the base name in `other-FHV-services-jan-july15`. Do you see anything that is a common abbreviation at the end of business names?

```sql
SELECT TOP 10 RIGHT(Base_Name, 5) AS 'Abbr'
FROM other_FHV_services_jan_aug_2015
```

ANSWER: INC is a common abbreviation.

The number of trips and the number of pickups are both stored as strings in `other-FHV-services-jan-july15`. This is because the data provided included spaces in some of the numbers. Write a query that returns the location of spaces in both of these columns.

```sql
SELECT CHARINDEX(' ', "Number_of_Trips") AS 'Space Location'
FROM other_FHV_services_jan_aug_2015
```

NOTE: There was common confusion on what column is "Number of Pickups" so I did not include this part of the query. The column names are Base\_Number, Base\_Name, Pick\_Up\_Date, Number\_of\_Trips, and Number\_of\_Vehicles.

## Date Functions

For ``other-FHV-services-jan-july15``, display the name of month of the pickup date and order in descending order. Is the result of this query what you expected?

```sql
SELECT DATENAME(month, "Pick_Up_Date") AS "Sorted Pickup Date"
FROM other_FHV_services_jan_aug_2015
ORDER BY "Sorted Pickup Date" DESC;
```

ANSWER: This is the result I am expecting, because it is sorting in descending alphabetical order while only including the months January through August.

Using `other-FHV-services-jan-july15`, write a query that returns the day number of the pickup date and month name and orders it in ascending order by day number.

```sql
SELECT DATENAME(month, "Pick_Up_Date") AS "Month", DATENAME(day, "Pick_Up_Date") AS "Day"
FROM other_FHV_services_jan_aug_2015
ORDER BY "Day";
```

## Aggregate Functions

Write a query that returns the most easterly Lyft pickup.

```sql
SELECT MAX(start_lng)
FROM lyft
```

Write a query the returns the most northerly Lyft pickup.

```sql
SELECT MIN(start_lat)
FROM lyft
```

Using ``other-FHV-services-jan-july15``, write a query that connects the the base number and the base name in a string that uses the following format: base\_number: base\_name.

```sql
SELECT CONCAT_WS(': ', Base_Number, Base_Name)
FROM other_FHV_services_jan_aug_2015
```
